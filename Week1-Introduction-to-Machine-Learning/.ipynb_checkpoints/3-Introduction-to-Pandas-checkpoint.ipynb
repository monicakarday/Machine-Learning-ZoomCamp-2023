### Data Frames

```python
data = [
    ['Nissan', 'Stanza', 1991, 138, 4, 'MANUAL', 'sedan', 2000],
    ['Hyundai', 'Sonata', 2017, None, 4, 'AUTOMATIC', 'Sedan', 27150],
    ['Lotus', 'Elise', 2010, 218, 4, 'MANUAL', 'convertible', 54990],
    ['GMC', 'Acadia',  2017, 194, 4, 'AUTOMATIC', '4dr SUV', 34450],
    ['Nissan', 'Frontier', 2017, 261, 6, 'MANUAL', 'Pickup', 32340],
]

columns = [
    'Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders',
    'Transmission Type', 'Vehicle_Style', 'MSRP'
]
```

```python
df = pd.DataFrame(data, columns=columns) #data can be a list of lists, each list is the observations
df
```

### Index & Accessing elements

```python
df.index
```

```python
df['Make'].index
```

```python
df.index = ['a','b','c','d','e'] #Updating index values
df
```

```python
df.loc['a'] # accessing individual rows by value of index
```

```python
df.iloc[0] # accessing individual rows by location of index
```

```python
df.reset_index() # The change won't affect the original dataframe
```

```python
df.reset_index(drop=True) # It doesn't create additional column for index values
```

```python
df 
```

### Filtering

```python
df['Make'] == 'Nissan' 
```

```python
df[df['Make'] == 'Nissan'] # Returns a dataframe only where the condition is True
```

```python
df[ (df['Make'] == 'Nissan') & (df['Year']>=2015) ]
```

### String operations

```python
'machine learning zoomcamp'.replace(" ", "_")
```

```python
df['Vehicle_Style'].str.lower() # This will not affect the original DataFrame
```

```python
df
```

```python
df['Vehicle_Style'] = df['Vehicle_Style'].str.replace(' ', '_').str.lower()
df
```

### Summarizing operations

```python
df.describe()
```

```python
df.describe().round(2)
```

```python
df.nunique() # Returns number of unique values in each column
```

### Missing values

```python
df.isnull() #Returns a complete dataframe of True where value is null and False where it is not
```

```python
df.isna().any() # Returns whether a particular column has a null value or not
```

```python
df.isnull().sum() # Returns the sum or total number of null values in a given column
```

### Grouping
